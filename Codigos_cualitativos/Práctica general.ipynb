## GRÁFICOS RT VS CORRECTAS/INCORRECTAS

```
## Gráfico de RT para las respuestas correctas e incorrectas, separado por TE
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = pd.read_csv("Datos_fila_20240927 (1).csv",index_col=["id"])
data=data.dropna()
data=data[data['resp'] != 'timeout']
data.head(100)
d1 = data.sort_values("id")

#d1
#data.isna().sum()
data
```

```


### PARA TE 100 MS

d_100_1 = data[(data["time"] == 100) & (data["correcta"] == 1)]
d_100_1_rt = d_100_1[["rt"]]  
d_100_1_rt.columns=["rt correctas"]   #datos de 100ms de TE, correctos
d_100_0 = data[(data["time"] == 100) & (data["correcta"] == 0)]   #datos de 100ms de TE, incorrectos
d_100_0_rt = d_100_0[["rt"]]
d_100_0_rt.columns=["rt incorrectas"]
d_100_0_rt = d_100_0[["rt"]].apply(lambda x: -x)
d_100_0_rt.columns=["rt incorrectas"]

d_100_1_rt_join = pd.concat([d_100_1_rt, d_100_0_rt], axis=0)
d_100_1_rt_join.plot(kind="hist",
                     alpha=0.5, 
                     bins=100,
                     xlabel = "RT (ms)",
                     ylabel = "Frecuencia",
                     title="Frecuencia de RT para TE=100ms",
                     color= ['gold','black'])
plt.savefig("RT_100_ms.png")




```

```
### PARA TE 200 MS

# Respuestas correctas para 200ms
d_200_1 = data[(data["time"] == 200) & (data["correcta"] == 1)]  
d_200_1_rt = d_200_1[["rt"]]
d_200_1_rt.columns = ["rt correctas"]
 # Respuestas incorrectas para 200ms
d_200_0 = data[(data["time"] == 200) & (data["correcta"] == 0)] 
d_200_0_rt = d_200_0[["rt"]].apply(lambda x: -x)
d_200_0_rt.columns = ["rt incorrectas"]

d_200_1_rt_join = pd.concat([d_200_1_rt, d_200_0_rt], axis=0)
d_200_1_rt_join.plot(kind="hist",
                     alpha=0.5, 
                     bins=100,
                     xlabel = "RT (ms)",
                     ylabel = "Frecuencia",
                     title="Frecuencia de RT para TE=200ms",
                     color= ['gold','black'])
plt.savefig("RT_200_ms.png")
```
